# Using loop

```python
df2.head()
```

```python
df2.columns
```

```python
import numpy as np

import warnings as wr
wr.filterwarnings ('ignore')

for column in df2.columns:
    if df2[column].dtype == np.number:
        continue
    df2[column] = le.fit_transform(df2[column])
```

# Another Way to avoid warnings

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for column in df3.columns:
    if is_numeric_dtype (df3[column]):
        continue
    else:
        df3[column] = le.fit_transform(df3[column])
```

# Using loop

# Ordinal Encoding

```python
df7.region.unique()
```

```python
unique_region = ['southwest', 'southeast', 'northwest', 'northeast']
from sklearn.preprocessing import OrdinalEncoder
ordinal = OrdinalEncoder(categories=[unique_region])
```

```python
ordinal
```

```python
encoded = ordinal.fit_transform (df7[['region']])
```

```python
en = pd.DataFrame(encoded, columns = ['region1'])
```

# Using Loop

```python
df7.head()
```

```python
columns = ['sex', 'smoker', 'region']

for column in columns:
    unique = df7[column].unique()
    df7[column] = OrdinalEncoder(categories=[unique]).fit_transform (df7[[column]])
```
